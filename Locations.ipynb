```python
locationlist = []
y=0
for x in speech['title']:
    places = GeoText(speech['title'][y])
    places2 = GeoText(speech['speech'][y].title())
    try:
        #First, attempt to extract location info from the title
        locationlist.append(str(mode(places.cities)))
    except:
        #Second, attempt to extract location info from the speech starting with excluding out any null observations
        if(len(places2.cities)==0):
            locationlist.append(str('Unknown'))
        else:
            # Third, remove all non-standard location terms
            for z in places2.cities[:]:
                if(z=='Of' or z=='Obama' or z=='Man' or z=='Deal' or z=='Much' or z=='Most' or z=='Bar' or z=='March' or z=='Taylor' or z=='Police' or z=='Date' or z=='George' or z=='Wedding' or z=='Beijing' or z=='Barletta'):
                    places2.cities.remove(z)
            # Fourth, if there are no more observations left, end with an unknown
            if(len(places2.cities)==0):
                locationlist.append(str('Unknown'))
            # Finally, return with full conditions
            else:
                locationlist.append(str(places2.cities[0]))
    y=y+1
```

```python
speech.assign(location=locationlist)
```
